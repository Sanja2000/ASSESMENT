**ENERGY EFFICIENCY IN INDUSTRIAL AUTOMATION**



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('/content/ENB2012_data.xlsx - Φύλλο1.csv') 

```

Renaming the column names

```python
data.rename(columns = {'X1':'Compactness'}, inplace = True)
data.rename(columns = {'X2':'Surface area'}, inplace = True)
data.rename(columns = {'X3':'Wall area'}, inplace = True)
data.rename(columns = {'X4':'Roof area'}, inplace = True)
data.rename(columns = {'X5':'Overall height'}, inplace = True)
data.rename(columns = {'X6':'Orientation'}, inplace = True)
data.rename(columns = {'X7':'Glazing area'}, inplace = True)
data.rename(columns = {'X8':'Glazing distribution'}, inplace = True)
data.rename(columns = {'Y1':'Heating load'}, inplace = True)
data.rename(columns = {'Y2':'Cooling load'}, inplace = True)
```

Visualizing the dataset

```python
data.head()
```

Intializing feature variables to x.
Intializing target variables to y.

```python
x = data[['Compactness', 'Surface area', 'Wall area', 'Roof area', 'Overall height', 'Orientation', 'Glazing area', 'Glazing distribution']].values
y = data[['Heating load', 'Cooling load']]
```

Importing the train_test_split and splitting the data into train and test

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

Scaling the data

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(x_train,y_train)
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

Importing the LinearRegression and fitting it to the train data

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train_scaled, y_train)
```

We predicting the x_test 

```python
y_pred = regressor.predict(x_test_scaled)
y_pred
```

**EVALUATION METRICS**

By importing metrics we find the MSE, MAE & R squared score

```python
from sklearn import metrics
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))
print('R2- SCORE:', metrics.r2_score(y_test, y_pred))
```

**FEATURE SELECTION**

Using RFE(Repeated Feature Elimination) we select the important feature

```python
from sklearn.feature_selection import RFE
estimator = LinearRegression(normalize = True)
selector = RFE(estimator, n_features_to_select=2, step=1)
selector = selector.fit(x, y)
list(zip(['Compactness', 'Surface area', 'Wall area', 'Roof area', 'Overall height', 'Orientation', 'Glazing area', 'Glazing distribution'],selector.support_))
```

We've found that Wall area and Roof area are the top most important feature
